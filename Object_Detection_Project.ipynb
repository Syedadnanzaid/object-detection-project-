```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split


DATA_DIR = "/kaggle/input/obj-detection2/data"
IMG_DIR = os.path.join(DATA_DIR, "training_images")
CSV_PATH = os.path.join(DATA_DIR, "train_solution_bounding_boxes (1).csv")
df = pd.read_csv(CSV_PATH)
df["image"] = df["image"].astype(str).str.strip()


N_total = 100
df = df.head(N_total)  

IMG_SIZE = (128, 128)


def preprocess_img_and_bbox(row):
    img_path = os.path.join(IMG_DIR, row["image"])
    img_orig = load_img(img_path)
    img_w, img_h = img_orig.size
    img = img_to_array(load_img(img_path, target_size=IMG_SIZE)) / 255.0
    
    xmin = float(row['xmin']) / img_w
    ymin = float(row['ymin']) / img_h
    xmax = float(row['xmax']) / img_w
    ymax = float(row['ymax']) / img_h
    return img, [xmin, ymin, xmax, ymax], img_w, img_h

imgs, boxes, ws, hs = [], [], [], []
for idx, row in df.iterrows():
    img, box, w, h = preprocess_img_and_bbox(row)
    imgs.append(img)
    boxes.append(box)
    ws.append(w)
    hs.append(h)
imgs = np.array(imgs, dtype=np.float32)
boxes = np.array(boxes, dtype=np.float32)


X_train, X_val, y_train, y_val = train_test_split(imgs, boxes, test_size=0.2, random_state=42)


model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation="relu", input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation="relu"),
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation="relu"),
    layers.Dense(4, activation="sigmoid")  
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()


history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=20,              
    batch_size=8,
    verbose=2
)


N_show = 3
fig, axs = plt.subplots(1, N_show, figsize=(8*N_show, 7))

for i in range(N_show):
    idx = i
    row = df.iloc[idx]
    img_path = os.path.join(IMG_DIR, row["image"])
    orig_img = load_img(img_path)
    orig_w, orig_h = orig_img.size
    img_resized = img_to_array(load_img(img_path, target_size=IMG_SIZE)) / 255.0
    img_input = np.expand_dims(img_resized, 0)

    pred_norm = model.predict(img_input)[0]
    pred_norm = np.clip(pred_norm, 0, 1)

    
    pred_box = [
        pred_norm[0] * orig_w,
        pred_norm[1] * orig_h,
        pred_norm[2] * orig_w,
        pred_norm[3] * orig_h
    ]
    gt_box = [
        float(row['xmin']),
        float(row['ymin']),
        float(row['xmax']),
        float(row['ymax'])
    ]

    ax = axs[i]
    ax.imshow(orig_img)
    ax.add_patch(patches.Rectangle((gt_box[0], gt_box[1]),
                                   gt_box[2] - gt_box[0], gt_box[3] - gt_box[1],
                                   linewidth=2, edgecolor="green", facecolor="none", label="Ground Truth"))
    ax.add_patch(patches.Rectangle((pred_box[0], pred_box[1]),
                                   pred_box[2] - pred_box[0], pred_box[3] - pred_box[1],
                                   linewidth=2, edgecolor="red", facecolor="none", label="Predicted"))
    ax.set_title(f"Prediction vs Ground Truth\n{row['image']}")
    ax.legend()

plt.tight_layout()
plt.show()
```
